```python
# try some probability distribution functions
tp = np.linspace(0,10,200)
ddc_pdf = sps.expon.pdf(x=tp,scale=ddc_t0_three.mean())
k1_pdf = sps.expon.pdf(x=tp,scale=k1_t0_three.mean())
k3_pdf = sps.expon.pdf(x=tp,scale=k3_t0_three.mean())
k5_pdf = sps.expon.pdf(x=tp,scale=k5_t0_three.mean())
```

```python
# lets estimate the parameters for the gamma function from the data
# first by maximum likelihood
def MLE_gamma(data):
    avg = data.mean()
    s = np.log(avg) - np.log(data).mean()
    k0 = (3 - s + np.sqrt((s -3)**2 +24*s))/12/s
#    print("initial guess:",k0)
    for i in range(4):
        k = k0 - (np.log(k0)-special.digamma(k0)-s)/(1/k0-special.polygamma(1,k0))
#        print("next guess:",k)
        k0=k
    return (k,avg/k)

kddc,theta_ddc = MLE_gamma(ddc_t0_three)
print(kddc,theta_ddc)
ddc_gamma = sps.gamma.pdf(x=tp,a=kddc,scale=theta_ddc)

kk1,theta_k1 = MLE_gamma(k1_t0_three)
print(kk1,theta_k1)
k1_gamma = sps.gamma.pdf(x=tp,a=kk1,scale=theta_k1)

kk3,theta_k3 = MLE_gamma(k3_t0_three)
print(kk3,theta_k3)
k3_gamma = sps.gamma.pdf(x=tp,a=kk3,scale=theta_k3)

kk5,theta_k5 = MLE_gamma(k5_t0_three)
print(kk5,theta_k5)
k5_gamma = sps.gamma.pdf(x=tp,a=kk5,scale=theta_k5)
```

```python
# let estimate gamma parameters by mean and variance
ddcvar = ddc_t0_three.std()**2
ddcmean = ddc_t0_three.mean()

ddc_theta = ddcvar/ddcmean
ddc_k = ddcmean/ddc_theta
print(ddc_k,ddc_theta)
ddc_gamma2 = sps.gamma.pdf(x=tp,a=ddc_k,scale=ddc_theta)

k1var = k1_t0_three.std()**2
k1mean = k1_t0_three.mean()

k1_theta = k1var/k1mean
k1_k = k1mean/k1_theta
print(k1_k,k1_theta)

k1_gamma2 = sps.gamma.pdf(x=tp,a=k1_k,scale=k1_theta)

k3var = k3_t0_three.std()**2
k3mean = k3_t0_three.mean()

k3_theta = k3var/k3mean
k3_k = k3mean/k3_theta
print(k3_k,k3_theta)

k3_gamma2 = sps.gamma.pdf(x=tp,a=k3_k,scale=k3_theta)

k5var = k5_t0_three.std()**2
k5mean = k5_t0_three.mean()

k5_theta = k5var/k5mean
k5_k = k5mean/k5_theta
print(k5_k,k5_theta)

k5_gamma2 = sps.gamma.pdf(x=tp,a=k5_k,scale=k5_theta)
```

```python
# lets also try a lognormal distribution
def MLE_lognormal(data):
    mu = np.log(data).mean()
    sigma2 = ((np.log(data)-mu)**2).mean()
    return (mu,np.sqrt(sigma2))

mu_ddc, sigma_ddc = MLE_lognormal(ddc_t0_three)
ddc_lognormal = sps.lognorm.pdf(x=tp,s=sigma_ddc,scale=np.exp(mu_ddc))

mu_k1, sigma_k1 = MLE_lognormal(k1_t0_three)
k1_lognormal = sps.lognorm.pdf(x=tp,s=sigma_k1,scale=np.exp(mu_k1))

mu_k3, sigma_k3 = MLE_lognormal(k3_t0_three)
k3_lognormal = sps.lognorm.pdf(x=tp,s=sigma_k3,scale=np.exp(mu_k3))

mu_k5, sigma_k5 = MLE_lognormal(k5_t0_three)
k5_lognormal = sps.lognorm.pdf(x=tp,s=sigma_k5,scale=np.exp(mu_k5))
```
