```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ddc_t0_two = np.load("ddc_t0_two.npy")
ddc_t0_three = np.load("ddc_t0_three.npy")
ddc_t0_cluster = np.load("ddc_t0_cluster.npy")
```

```python
plt.figure()
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(3, sharex=True)
axarr[0].hist(ddc_t0_two, bins=20,normed=True,label='two stage model')
axarr[0].set_title('t0 distributions')
axarr[0].legend()
axarr[1].hist(ddc_t0_three,bins=20,color='g',normed=True,label='three stage model')
axarr[1].legend()
axarr[2].hist(ddc_t0_cluster,bins=20,color='r',normed=True,label='cluster analysis')
axarr[2].legend()
```

```python
lipo_t0_two = np.load("lipo_t0_two.npy")
lipo_t0_three = np.load("lipo_t0_three.npy")
lipo_t0_cluster = np.load("lipo_t0_cluster.npy")
```

```python
plt.figure()
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(3, sharex=True)
axarr[0].hist(lipo_t0_two, bins=20,normed=True,label='two stage model')
axarr[0].set_title('t0 distributions')
axarr[0].legend()
axarr[1].hist(lipo_t0_three,bins=20,color='g',normed=True,label='three stage model')
axarr[1].legend()
axarr[2].hist(lipo_t0_cluster,bins=20,color='r',normed=True,label='cluster analysis')
axarr[2].legend()
```

```python
lipo2_t0_two = np.load("lipo2_t0_two.npy")
lipo2_t0_three = np.load("lipo2_t0_three.npy")
lipo2_t0_cluster = np.load("lipo2_t0_cluster.npy")
```

```python
lipo_t0_three_all = np.concatenate((lipo_t0_three,lipo2_t0_three))
lipo_t0_two_all = np.concatenate((lipo_t0_two,lipo2_t0_two))
lipo_t0_cluster_all = np.concatenate((lipo_t0_cluster,lipo2_t0_cluster))
```

```python
plt.figure()
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].hist(ddc_t0_three, bins=20,normed=True,label='three stage DDC\nt0 mean = 1.04 hours')
axarr[0].set_title('t0 distributions')
axarr[0].legend()
axarr[1].hist(lipo_t0_three_all,bins=30,color='g',normed=True,label='three stage lipo\nt0 mean = 2.66 hours')
axarr[1].legend()
plt.xlabel("delay time in hours")
print(lipo_t0_three_all.mean())
print(ddc_t0_three.mean())
```

```python
plt.figure()
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].hist(ddc_t0_two, bins=20,normed=True,label='two stage DDC\nt0 mean = 2.08 hours')
axarr[0].set_title('t0 distributions')
axarr[0].legend()
axarr[1].hist(lipo_t0_two_all,bins=30,color='g',normed=True,label='two stage lipo\nt0 mean = 3.56 hours')
axarr[1].legend()
plt.xlabel("delay time in hours")
print(lipo_t0_two_all.mean())
print(ddc_t0_two.mean())
```

```python
plt.figure()
# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].hist(ddc_t0_cluster, bins=20,normed=True,label='cluster DDC\nt0 mean = 1.23 hours')
axarr[0].set_title('t0 distributions')
axarr[0].legend()
axarr[1].hist(lipo_t0_cluster_all,bins=30,color='g',normed=True,label='cluster lipo\nt0 mean = 2.21 hours')
axarr[1].legend()
plt.xlabel("delay time in hours")
print(lipo_t0_cluster_all.mean())
print(ddc_t0_cluster.mean())
```
